```python
obs_BODY_HEIGHT = pd.to_numeric(observations[(observations['DESCRIPTION'] == 'Body Height' )]['VALUE'])
obs_BODY_WEIGHT = pd.to_numeric(observations[(observations['DESCRIPTION'] == 'Body Weight' )]['VALUE'])
obs_BMI = pd.to_numeric(observations[(observations['DESCRIPTION'] == 'Body Mass Index' )]['VALUE'])
obs_DBP = pd.to_numeric(observations[(observations['DESCRIPTION'] == 'Diastolic Blood Pressure' )]['VALUE'])
obs_SBP = pd.to_numeric(observations[(observations['DESCRIPTION'] == 'Systolic Blood Pressure' )]['VALUE'])
```

```python
print('**********----------------------------------------------------------**********')
print('Statistics about BODY Height in cm : {}'.format(obs_BODY_HEIGHT.describe()))
print('**********----------------------------------------------------------**********')
print('Statistics about BODY Weight in kg : {}'.format(obs_BODY_WEIGHT.describe()))
print('**********----------------------------------------------------------**********')
print('Statistics about  BMI in kg/m2 : {}'.format(obs_BMI.describe()))
print('**********----------------------------------------------------------**********')
print('Statistics about Diastolic Blood Pressure in mm[Hg] : {}'.format(obs_DBP.describe()))
print('**********----------------------------------------------------------**********')
print('Satistics about Systolic Blood Pressure in mm[Hg] : {}'.format(obs_SBP.describe()))
print('**********----------------------------------------------------------**********')
```

```python
#merging temporary (patients and conditions) to allergies
patient_cond_allg = pd.merge(temp,allergies,left_on = 'Id', right_on = 'PATIENT')
```

```python
patient_cond_allg.head()
```

```python
patient_cond_allg.columns
```

```python
patient_cond_allg.drop(['BIRTHDATE', 'DEATHDATE', 'SSN', 'DRIVERS', 'PASSPORT', 
                        'PREFIX','FIRST', 'LAST', 'SUFFIX', 'MAIDEN', 'MARITAL',
                        'BIRTHPLACE', 'ADDRESS','COUNTY', 'ZIP','LAT', 'LON', 
                        'HEALTHCARE_EXPENSES', 'HEALTHCARE_COVERAGE', 'START_x',
                        'STOP_x', 'PATIENT_x', 'ENCOUNTER_x', 'CODE_x','START_y', 
                        'STOP_y', 'PATIENT_y', 'ENCOUNTER_y', 'CODE_y'],
                       axis = 1, inplace = True)
```
